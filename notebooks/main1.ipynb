<h1>Match Analysis<h1>

```python
# Bat First vs Chase Wins

# Create Batting First Team
matches['bat_first_team'] = matches.apply(
    lambda x: x['team1']
    if x['toss_decision'] == 'bat'
    else x['team2'],
    axis=1
)

# Create Chasing Team
matches['chasing_team'] = matches.apply(
    lambda x: x['team2']
    if x['toss_decision'] == 'bat'
    else x['team1'],
    axis=1
)

# Find Bat First Wins
bat_first_wins = (matches['winner'] == matches['bat_first_team']).sum()

print(bat_first_wins)


# Find Chase Wins
chase_wins = (matches['winner'] == matches['chasing_team']).sum()

print(chase_wins)

# Calculate Percentages
total_matches = len(matches)

bat_first_pct = round(bat_first_wins / total_matches * 100,2)

chase_pct = round(chase_wins / total_matches * 100,2)

print("Bat First Wins:", bat_first_pct)
print("Chase Wins:", chase_pct)
```

```python
summary = pd.DataFrame({
    'Strategy':['Bat First','Chase'],
    'Wins':[bat_first_wins,chase_wins],
    'Percentage':[bat_first_pct,chase_pct]
})

summary
```

<h4>Best Finisher Batsman



```python
# Filter Death Over
death_overs = deliveries[deliveries['over'] >= 16]

# Calculate Runs and Ball Faced
finisher_stats = death_overs.groupby('batsman').agg(
    Runs=('batsman_runs', 'sum'),
    Balls=('batsman_runs', 'count')
)

# Calculate Strike Rate
finisher_stats['Strike Rate'] = (finisher_stats['Runs'] / finisher_stats['Balls']) * 100

# Filter players who faced at least 100 balls
finisher_stats = finisher_stats[finisher_stats['Balls'] >= 100]

# Create Finisher Score
# finisher_stats['Finisher Score'] = (
#     finisher_stats['Runs'] * 0.7
#     +
#     finisher_stats['Strike Rate'] * 0.3
# )

# Rank Best Finishers
best_finishers = finisher_stats.sort_values('Runs',ascending=False)

pd.DataFrame(best_finishers).head(10)
```

<h4>Best Powerplay Batsman

```python
# Filter Powerplay Overs
powerplay = deliveries[deliveries['over'] <= 6]

# Calculate Runs
powerplay_runs = powerplay.groupby('batsman')['batsman_runs'].sum()
powerplay_runs.sort_values(ascending=False).head(10)

# Calculate Balls Faced
balls_faced = powerplay.groupby('batsman').size()

# Create Batting Summary
powerplay_stats = pd.DataFrame({
    'Runs': powerplay.groupby('batsman')['batsman_runs'].sum(),
    'Balls': powerplay.groupby('batsman').size()
})

# Calculate Strike Rate
powerplay_stats['Strike Rate'] = (
    powerplay_stats['Runs']
    /
    powerplay_stats['Balls']
) * 100

# Runs Score
powerplay_stats['Runs Score'] = (
    powerplay_stats['Runs']
    /
    powerplay_stats['Runs'].max()
) * 100

# Strike Rate Score
powerplay_stats['SR Score'] = (
    powerplay_stats['Strike Rate']
    /
    powerplay_stats['Strike Rate'].max()
) * 100

# Impact Score
powerplay_stats['Impact Score'] = (
    powerplay_stats['Runs Score'] * 0.6
    +
    powerplay_stats['SR Score'] * 0.4
)

# Filter players with at least 100 balls
powerplay_stats = powerplay_stats[
    powerplay_stats['Balls'] >= 200
]
```

<h1>MVP (Most Valuable Player) Score

```python
# Calculate Batting Runs
runs = deliveries.groupby('batsman')['batsman_runs'].sum()
runs = runs.reset_index()
runs.columns = ['Player', 'Runs']

# Calculate Wickets
wickets = deliveries[
    deliveries['player_dismissed'].notnull()
]

wickets = wickets.groupby('bowler').size().reset_index()

wickets.columns = ['Player', 'Wickets']

# Calculate Player of Match Awards
awards = matches.groupby(
    'player_of_match'
).size().reset_index()

awards.columns = ['Player', 'Awards']

# Merge Everything
mvp = runs.merge(
    wickets,
    on='Player',
    how='outer'
)

mvp = mvp.merge(
    awards,
    on='Player',
    how='outer'
)

mvp.fillna(0, inplace=True)

# Normalize Values
mvp['Runs_Score'] = (
    mvp['Runs']
    /
    mvp['Runs'].max()
) * 100

mvp['Wicket_Score'] = (
    mvp['Wickets']
    /
    mvp['Wickets'].max()
) * 100

mvp['Award_Score'] = (
    mvp['Awards']
    /
    mvp['Awards'].max()
) * 100

# Create MVP Score
mvp['MVP_Score'] = (
    mvp['Runs_Score'] * 0.5
    +
    mvp['Wicket_Score'] * 0.3
    +
    mvp['Award_Score'] * 0.2
)

# Rank Players
mvp = mvp.sort_values(
    'MVP_Score',
    ascending=False
)

mvp.head(10)
```
